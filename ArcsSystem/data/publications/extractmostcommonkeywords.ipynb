```python
import csv
from xml.sax.saxutils import escape
from collections import Counter
```

```python
def parsecsv():
    with open('ARCSLiterature.csv', 'r') as thefile:
        data = csv.DictReader(thefile, delimiter=',', quotechar='"')
        datadict = {}
        for row in data:
            if escape(row['Title']) not in datadict:
                datadict[row['Title']] = [row['Author'],
                                                 row['Publication Year'],
                                                 row['Publication Title'],
                                                 row['Issue'],
                                                 row['Volume'],
                                                 row['DOI'],
                                                 row['Abstract Note'],
                                                 row['Manual Tags']
                                          ]

            else:
                continue
        return datadict
```

```python
keyworddict = {}

kwordlist = []

counter = 1

for k, v in parsecsv().items():
    keywordlist = v[7].split("; ")
    for x in keywordlist:
        loweredkeyword = escape(x.lower())
        if len(loweredkeyword) > 1:
            kwordlist.append(loweredkeyword)
            if loweredkeyword not in keyworddict:
                keyworddict[loweredkeyword] = counter
                counter += 1
           
        
del keyworddict["newtech"]
for k in kwordlist:
    if k == "newtech":
        kwordlist.remove(k)
```

```python


print(Counter(kwordlist).most_common(500))
  
```

```python
orderedkwlist = []
for c in Counter(kwordlist).most_common():
    orderedkwlist.append(c[0])  
print(len(orderedkwlist))
print(orderedkwlist)
```

```python
with open("top500keywords.csv", "w") as csvfile:
    kwriter = csv.writer(csvfile)
    for k in Counter(kwordlist).most_common(500):
        kwriter.writerow([k[0]])
    
```
